14. Write a program that takes a user’s full name (with spaces), converts it to both uppercase and lowercase, uses strip() and split(), and then rejoins the words.

```python
name = input("Enter your name: ")
name= name.strip()          
words = name.split()         
join = " ".join(words)     
print(name)
print(words)
print(join)
print("Uppercase:", join.upper())
print("Lowercase:", join.lower())
```
