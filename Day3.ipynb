1. Ask the user for a string and display its reverse using slicing.

```python
name = input("Enter any name for reverse: ")
reverse_name = name[::-1]
print("Reverse:", reverse_name)
```

2.  Input a string and a character. Use  .count() to find  how many times it appears. 


```python
s = input("Enter a string: ")
c = input("Enter a character: ")
count = s.count(c)
print("The character appears", count, "times.")
```

3.  Input a sentence and convert it into a word list  using .split()

```python
sentence = input("Enter a sentence: ")
list = sentence.split()
print("Word list:", list)
```

4.  Create a list of words and join them into a  single string using .join() 

```python
words = ["I", "am", "Shreya", "Rastogi"]
sentence = " ".join(words)
print("Joined string:", sentence)
```

5. Input 5 numbers, store them in a list, and sort them using .sort() 

```python
numbers = input("Enter 5 numbers separated by space: ").split()
numbers.sort()
print("Sorted list:", numbers)
```

6. From a list of numbers, print only even numbers.

```python
numbers = [10, 3, 4, 7, 12, 15, 20]
for n in numbers:
    if n % 2 == 0:
        print(n)
```

7. Create a nested list with a student's name and marks. Print name, all marks, and one specific mark

```python
student = ["Shreya", [85, 90, 78, 92]]
print("Name:", student[0])
print("All marks:", student[1])
print("Second mark:", student[1][1])
```
