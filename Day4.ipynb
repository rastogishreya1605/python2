1. Create a tuple of 6 fruits and print: first, last, and middle item.

```python
fruits = ("apple", "banana", "mango", "orange", "grapes", "papaya")
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])
print("Middle fruits:", fruits[2], "and", fruits[3])
```

2.  Create a tuple (name , age , course ) and unpack values.

```python
student = ("Shreya", 20, "BCA")
name, age, course = student
print("Name:", name)
print("Age:", age)
print("Course:", course)
```

3.  Create two sets and print: union, intersection, and difference.

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
print("Difference:", set1.difference(set2))
```

4. Convert a list with repeated numbers to a set and print the set.

```python
numbers = [1, 2, 2, 3, 4, 4, 5, 5, 5]
unique= set(numbers)
print("Unique numbers:", unique)
```

5. Create a dictionary and print keys and values separately.

```python
student = {"name": "Shreya", "age": 20, "course": "BCA"}
print("Keys:", student.keys())
print("Values:", student.values())
```

6.  Add "grade" and update "marks" value.

```python
student = { "name": "Shreya", "marks": 85, "course": "BCA"}
student["grade"] = "A"
student["marks"] = 90
print(student)
```

7. Print all key-value pairs using a loop.

```python
student = { "name": "Shreya", "age": 20, "course": "BCA", "marks": 90, "grade": "A"}
for key, value in student.items():
    print(key, ":", value)
```
