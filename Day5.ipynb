1. Take a number and check whether it is positive, negative, or zero.

```python
n = int(input("Enter a number: "))
if n > 0:
    print("The number is Positive")
elif n < 0:
    print("The number is Negative")
else:
    print("The number is Zero")
```

2. Based on marks input, print grade A/B/C/Fail.

```python
marks = int(input("Enter your marks: "))
if marks >= 90:
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

3. Print numbers from 1 to 20 using a loop.

```python
for i in range(1, 21):
    print(i)
```

4. Print numbers 10 → 1 using range() .

```python
for i in range(10, 0, -1):
    print(i)
```
